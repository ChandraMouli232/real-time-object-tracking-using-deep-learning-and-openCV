```python
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox

# Ensure you have correct paths to the YOLOv3 weights and config files.
# If you're using default cvlib detection, it should automatically download them.
# Otherwise, make sure to specify the full path if needed.

# Function to detect and draw bounding boxes
def detect_and_draw_box(img, confidence=0.6):
    # Detect objects in the image
    bbox, label, conf = cv.detect_common_objects(img, confidence=confidence)
    # Draw bounding boxes
    output_image = draw_bbox(img, bbox, label, conf)
    return output_image

# Capture video from webcam
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()
    if not ret:
        break

    # Process each frame to detect and draw bounding boxes
    output_frame = detect_and_draw_box(frame)

    # Show the processed frame in a window
    cv2.imshow('Object Detection', output_frame)

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
vid.release()
cv2.destroyAllWindows()
```
